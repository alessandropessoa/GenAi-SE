# Contexto

Proposta de Framework: IFOR-GenAI (Índice de Faixa Operacional e Rangeabilidade para GenAI)
1. O que está sendo proposto? (Visão Geral)
Estamos propondo a criação e implementação de um framework de métricas avançado, o IFOR-GenAI. Inspirado em conceitos consolidados de instrumentação e controle de processos industriais, este framework avalia de forma objetiva, contínua e automatizada a eficácia, confiabilidade e o retorno sobre o investimento (ROI) de ferramentas de Inteligência Artificial Generativa no contexto real da engenharia de software. O objetivo é transformar a maneira como medimos e gerenciamos o impacto da IA, passando de impressões subjetivas para uma análise de engenharia precisa.

2. Funcionalidades e Metodologia (Como funciona?)
O IFOR-GenAI não é apenas uma métrica, mas um sistema completo com funcionalidades distintas que operam em conjunto para gerar insights profundos. A implementação é realizada através de uma metodologia robusta e automatizada:

Coleta de Dados Não Intrusiva: O sistema se integra diretamente ao ambiente de desenvolvimento (IDE) dos praticantes de software, capturando de forma automática e em segundo plano os dados essenciais de cada interação com a IA: o prompt inicial do desenvolvedor, o código gerado pela IA e a versão final do código após a revisão e modificação humana.

Análise Semântica Avançada (Além do Code Churn): Em vez de uma simples comparação de texto, o "motor" do IFOR-GenAI utiliza embeddings de código para analisar a similaridade semântica entre a sugestão da IA e o resultado final. Isso permite diferenciar uma refatoração inteligente (onde a ideia da IA foi boa, mas a implementação foi melhorada) de uma sugestão inútil que foi completamente descartada, resultando em um Índice de Qualidade Avançado (IQA) muito mais preciso.

Classificação de Tarefas por IA: Utilizando a mesma tecnologia de embeddings, o sistema classifica automaticamente cada tarefa com base no prompt do desenvolvedor, categorizando-a por domínio (ex: backend, testes, frontend) e complexidade inferida. Isso elimina a necessidade de anotação manual e permite uma análise em larga escala.

Painel de Controle Interativo (A Faixa de Operacional Ideal): Os dados processados alimentam um dashboard visual, o Mapa da Faixa de Operação Ideal (FOI). Este mapa de calor mostra de forma clara e imediata os pontos fortes (zonas de alto IQA) e os pontos fracos (zonas de baixo IQA) da ferramenta de IA para os diferentes tipos de tarefa.

Análise de Versatilidade (Índice de Rangeabilidade): A partir do mapa, o sistema calcula o Índice de Rangeabilidade (R), um KPI chave que quantifica a versatilidade da IA. O limiar de qualidade para este cálculo é ajustável dinamicamente, permitindo que gestores façam diferentes perguntas aos dados, como "Onde a IA é uma especialista autônoma?" (limite alto) versus "Onde ela serve ao menos como assistente de rascunho?" (limite baixo).

Geração de Relatórios por LLM: Como funcionalidade final, os dados estruturados e as métricas calculadas são enviados para a API do Gemini, que gera um relatório narrativo completo, com sumário executivo, diagnóstico aprofundado e recomendações estratégicas, transformando dados brutos em inteligência acionável.

3. Relevância Estratégica e Tática (Por que é importante?)
A implementação do IFOR-GenAI é fundamental para qualquer organização que busca utilizar a IA Generativa de forma séria, segura e eficiente. Sua relevância se manifesta em três níveis:

Relevância Estratégica (Para a Gestão):

ROI Comprovado: Permite medir o verdadeiro retorno sobre o investimento em licenças de IA, justificando custos com base em ganhos de eficiência mensuráveis.

Decisões de Ferramental: Fornece dados objetivos para comparar diferentes ferramentas de IA no mercado e decidir qual se adapta melhor às necessidades da equipe.

Vantagem Competitiva: Permite que a organização otimize seus fluxos de trabalho com IA de forma mais rápida e segura que a concorrência.

Relevância Tática (Para as Equipes de Engenharia):

Mitigação de Riscos: Identifica as "zonas de perigo" onde o código gerado pela IA é propenso a erros, permitindo a implementação de políticas de revisão de código mais rigorosas nesses cenários.

Otimização do Fluxo de Trabalho: Guia os desenvolvedores sobre quando e como usar a IA para obter o máximo benefício, evitando a frustração de usá-la em tarefas onde ela não é eficaz.

Aceleração da Capacitação: Serve como uma ferramenta de feedback para treinar desenvolvedores (especialmente os mais novos) sobre como interagir com a IA de forma produtiva e segura.

Relevância Técnica (Para a Qualidade de Software):

Melhoria Contínua da Qualidade: Ao entender as fraquezas da IA, as equipes podem focar a atenção humana onde ela é mais necessária, melhorando a qualidade geral do código.

Criação de um Ciclo de Feedback: Os dados coletados podem ser usados para refinar as práticas de engenharia de prompts da equipe, criando um ciclo virtuoso de melhoria na interação humano-IA.

Em suma, o IFOR-GenAI propõe tratar a IA não como uma "caixa-preta" mágica, mas como uma ferramenta de engenharia de alta precisão que deve ser calibrada, monitorada e gerenciada para garantir performance, segurança e o máximo impacto positivo nos resultados da organização.

 # **O MOTOR DE ANÁLISE (FUNÇÕES)**

```python
# Inicializa o vetorizador que simulará nossos embeddings
# TfidfVectorizer converte texto em vetores numéricos.
vectorizer = TfidfVectorizer()

def calcular_code_churn(codigo_original, codigo_final):
    """Calcula o churn com base nas linhas adicionadas/removidas."""
    seq_matcher = difflib.SequenceMatcher(None, codigo_original.splitlines(), codigo_final.splitlines())
    mudancas = 0
    for tag, i1, i2, j1, j2 in seq_matcher.get_opcodes():
        if tag != 'equal':
            mudancas += max(i2 - i1, j2 - j1)

    total_linhas = len(codigo_final.splitlines())
    return mudancas / total_linhas if total_linhas > 0 else 0

def classificar_tarefa(prompt_embedding, centroides):
    """Classifica um prompt encontrando a categoria mais similar."""
    similaridades = cosine_similarity(prompt_embedding, list(centroides.values()))
    categoria_idx = np.argmax(similaridades)
    return list(centroides.keys())[categoria_idx]

def calcular_iqa(codigo_ia, codigo_humano, embedding_ia, embedding_humano, alfa=0.5):
    """Calcula o Índice de Qualidade Avançado."""
    churn = calcular_code_churn(codigo_ia, codigo_humano)
    similaridade_semantica = cosine_similarity(embedding_ia, embedding_humano)[0][0]

    iqa = (alfa * (1 - churn)) + ((1 - alfa) * similaridade_semantica)
    return iqa, churn, similaridade_semantica
```

```python
vectorizer = TfidfVectorizer()

def calcular_code_churn(codigo_original, codigo_final):
    seq_matcher = difflib.SequenceMatcher(None, codigo_original.splitlines(), codigo_final.splitlines())
    mudancas = 0
    for tag, i1, i2, j1, j2 in seq_matcher.get_opcodes():
        if tag != 'equal':
            mudancas += max(i2 - i1, j2 - j1)
    total_linhas = len(codigo_final.splitlines())
    return mudancas / total_linhas if total_linhas > 0 else 0

# <-- MUDANÇA: A função agora recebe a matriz empilhada e a lista de chaves
def classificar_tarefa(prompt_embedding, matriz_centroides, chaves_categorias):
    """Classifica um prompt encontrando a categoria mais similar."""
    similaridades = cosine_similarity(prompt_embedding, matriz_centroides)
    categoria_idx = np.argmax(similaridades)
    return chaves_categorias[categoria_idx]

def calcular_iqa(codigo_ia, codigo_humano, embedding_ia, embedding_humano, alfa=0.5):
    churn = calcular_code_churn(codigo_ia, codigo_humano)
    similaridade_semantica = cosine_similarity(embedding_ia, embedding_humano)[0][0]
    iqa = (alfa * (1 - churn)) + ((1 - alfa) * similaridade_semantica)
    return iqa, churn, similaridade_semantica
```

# O PROCESSAMENTO

```python

print("🚀 Iniciando processamento do IFOR-GenAI...\n")
```

```python

# --- Preparação do Classificador ---
categorias_definidas = {
    ("Testes", "Baixa"): "criar teste unitário simples",
    ("Backend", "Média"): "gerar endpoint de api flask",
    ("Backend", "Alta"): "refatorar módulo de pagamento de alta performance",
    ("Utils", "Baixa"): "criar função de validação de email"
}
vectorizer.fit(list(categorias_definidas.values()) + [d['prompt'] for d in dados_interacoes])

centroides = {cat: vectorizer.transform([prompt_exemplo]) for cat, prompt_exemplo in categorias_definidas.items()}

# <-- MUDANÇA: Empilhamos os vetores dos centroides em uma única matriz
chaves_categorias = list(centroides.keys())
matriz_centroides = vstack(list(centroides.values()))

# --- Processamento das Interações ---
resultados = []
for interacao in dados_interacoes:
    print(f"--- Processando Interação ID: {interacao['id']} ---")

    prompt_embedding = vectorizer.transform([interacao['prompt']])
    embedding_ia = vectorizer.transform([interacao['codigo_ia']])
    embedding_humano = vectorizer.transform([interacao['codigo_humano']])

    # <-- MUDANÇA: Chamada da função com os novos argumentos
    categoria_tarefa = classificar_tarefa(prompt_embedding, matriz_centroides, chaves_categorias)
    print(f"Tarefa classificada como: {categoria_tarefa}")

    iqa, churn, ss = calcular_iqa(
        interacao['codigo_ia'], interacao['codigo_humano'],
        embedding_ia, embedding_humano
    )
    print(f"Code Churn: {churn:.2f} | Similaridade Semântica: {ss:.2f} | IQA Final: {iqa:.2f}")

    resultados.append({
        "categoria": categoria_tarefa[0],
        "complexidade": categoria_tarefa[1],
        "iqa": iqa
    })
print("\n✅ Processamento concluído.\n")
```
