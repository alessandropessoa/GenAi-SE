# Contexto

Proposta de Framework: IFOR-GenAI (√çndice de Faixa Operacional e Rangeabilidade para GenAI)
1. O que est√° sendo proposto? (Vis√£o Geral)
Estamos propondo a cria√ß√£o e implementa√ß√£o de um framework de m√©tricas avan√ßado, o IFOR-GenAI. Inspirado em conceitos consolidados de instrumenta√ß√£o e controle de processos industriais, este framework avalia de forma objetiva, cont√≠nua e automatizada a efic√°cia, confiabilidade e o retorno sobre o investimento (ROI) de ferramentas de Intelig√™ncia Artificial Generativa no contexto real da engenharia de software. O objetivo √© transformar a maneira como medimos e gerenciamos o impacto da IA, passando de impress√µes subjetivas para uma an√°lise de engenharia precisa.

2. Funcionalidades e Metodologia (Como funciona?)
O IFOR-GenAI n√£o √© apenas uma m√©trica, mas um sistema completo com funcionalidades distintas que operam em conjunto para gerar insights profundos. A implementa√ß√£o √© realizada atrav√©s de uma metodologia robusta e automatizada:

Coleta de Dados N√£o Intrusiva: O sistema se integra diretamente ao ambiente de desenvolvimento (IDE) dos praticantes de software, capturando de forma autom√°tica e em segundo plano os dados essenciais de cada intera√ß√£o com a IA: o prompt inicial do desenvolvedor, o c√≥digo gerado pela IA e a vers√£o final do c√≥digo ap√≥s a revis√£o e modifica√ß√£o humana.

An√°lise Sem√¢ntica Avan√ßada (Al√©m do Code Churn): Em vez de uma simples compara√ß√£o de texto, o "motor" do IFOR-GenAI utiliza embeddings de c√≥digo para analisar a similaridade sem√¢ntica entre a sugest√£o da IA e o resultado final. Isso permite diferenciar uma refatora√ß√£o inteligente (onde a ideia da IA foi boa, mas a implementa√ß√£o foi melhorada) de uma sugest√£o in√∫til que foi completamente descartada, resultando em um √çndice de Qualidade Avan√ßado (IQA) muito mais preciso.

Classifica√ß√£o de Tarefas por IA: Utilizando a mesma tecnologia de embeddings, o sistema classifica automaticamente cada tarefa com base no prompt do desenvolvedor, categorizando-a por dom√≠nio (ex: backend, testes, frontend) e complexidade inferida. Isso elimina a necessidade de anota√ß√£o manual e permite uma an√°lise em larga escala.

Painel de Controle Interativo (A Faixa de Operacional Ideal): Os dados processados alimentam um dashboard visual, o Mapa da Faixa de Opera√ß√£o Ideal (FOI). Este mapa de calor mostra de forma clara e imediata os pontos fortes (zonas de alto IQA) e os pontos fracos (zonas de baixo IQA) da ferramenta de IA para os diferentes tipos de tarefa.

An√°lise de Versatilidade (√çndice de Rangeabilidade): A partir do mapa, o sistema calcula o √çndice de Rangeabilidade (R), um KPI chave que quantifica a versatilidade da IA. O limiar de qualidade para este c√°lculo √© ajust√°vel dinamicamente, permitindo que gestores fa√ßam diferentes perguntas aos dados, como "Onde a IA √© uma especialista aut√¥noma?" (limite alto) versus "Onde ela serve ao menos como assistente de rascunho?" (limite baixo).

Gera√ß√£o de Relat√≥rios por LLM: Como funcionalidade final, os dados estruturados e as m√©tricas calculadas s√£o enviados para a API do Gemini, que gera um relat√≥rio narrativo completo, com sum√°rio executivo, diagn√≥stico aprofundado e recomenda√ß√µes estrat√©gicas, transformando dados brutos em intelig√™ncia acion√°vel.

 # **O MOTOR DE AN√ÅLISE (FUN√á√ïES)**

```python
# Inicializa o vetorizador que simular√° nossos embeddings
# TfidfVectorizer converte texto em vetores num√©ricos.
vectorizer = TfidfVectorizer()

def calcular_code_churn(codigo_original, codigo_final):
    """Calcula o churn com base nas linhas adicionadas/removidas."""
    seq_matcher = difflib.SequenceMatcher(None, codigo_original.splitlines(), codigo_final.splitlines())
    mudancas = 0
    for tag, i1, i2, j1, j2 in seq_matcher.get_opcodes():
        if tag != 'equal':
            mudancas += max(i2 - i1, j2 - j1)

    total_linhas = len(codigo_final.splitlines())
    return mudancas / total_linhas if total_linhas > 0 else 0

def classificar_tarefa(prompt_embedding, centroides):
    """Classifica um prompt encontrando a categoria mais similar."""
    similaridades = cosine_similarity(prompt_embedding, list(centroides.values()))
    categoria_idx = np.argmax(similaridades)
    return list(centroides.keys())[categoria_idx]

def calcular_iqa(codigo_ia, codigo_humano, embedding_ia, embedding_humano, alfa=0.5):
    """Calcula o √çndice de Qualidade Avan√ßado."""
    churn = calcular_code_churn(codigo_ia, codigo_humano)
    similaridade_semantica = cosine_similarity(embedding_ia, embedding_humano)[0][0]

    iqa = (alfa * (1 - churn)) + ((1 - alfa) * similaridade_semantica)
    return iqa, churn, similaridade_semantica
```

```python
vectorizer = TfidfVectorizer()

def calcular_code_churn(codigo_original, codigo_final):
    seq_matcher = difflib.SequenceMatcher(None, codigo_original.splitlines(), codigo_final.splitlines())
    mudancas = 0
    for tag, i1, i2, j1, j2 in seq_matcher.get_opcodes():
        if tag != 'equal':
            mudancas += max(i2 - i1, j2 - j1)
    total_linhas = len(codigo_final.splitlines())
    return mudancas / total_linhas if total_linhas > 0 else 0

# <-- MUDAN√áA: A fun√ß√£o agora recebe a matriz empilhada e a lista de chaves
def classificar_tarefa(prompt_embedding, matriz_centroides, chaves_categorias):
    """Classifica um prompt encontrando a categoria mais similar."""
    similaridades = cosine_similarity(prompt_embedding, matriz_centroides)
    categoria_idx = np.argmax(similaridades)
    return chaves_categorias[categoria_idx]

def calcular_iqa(codigo_ia, codigo_humano, embedding_ia, embedding_humano, alfa=0.5):
    churn = calcular_code_churn(codigo_ia, codigo_humano)
    similaridade_semantica = cosine_similarity(embedding_ia, embedding_humano)[0][0]
    iqa = (alfa * (1 - churn)) + ((1 - alfa) * similaridade_semantica)
    return iqa, churn, similaridade_semantica
```

# O PROCESSAMENTO

```python

print("üöÄ Iniciando processamento do IFOR-GenAI...\n")
```

```python

# --- Prepara√ß√£o do Classificador ---
categorias_definidas = {
    ("Testes", "Baixa"): "criar teste unit√°rio simples",
    ("Backend", "M√©dia"): "gerar endpoint de api flask",
    ("Backend", "Alta"): "refatorar m√≥dulo de pagamento de alta performance",
    ("Utils", "Baixa"): "criar fun√ß√£o de valida√ß√£o de email"
}
vectorizer.fit(list(categorias_definidas.values()) + [d['prompt'] for d in dados_interacoes])

centroides = {cat: vectorizer.transform([prompt_exemplo]) for cat, prompt_exemplo in categorias_definidas.items()}

# <-- MUDAN√áA: Empilhamos os vetores dos centroides em uma √∫nica matriz
chaves_categorias = list(centroides.keys())
matriz_centroides = vstack(list(centroides.values()))

# --- Processamento das Intera√ß√µes ---
resultados = []
for interacao in dados_interacoes:
    print(f"--- Processando Intera√ß√£o ID: {interacao['id']} ---")

    prompt_embedding = vectorizer.transform([interacao['prompt']])
    embedding_ia = vectorizer.transform([interacao['codigo_ia']])
    embedding_humano = vectorizer.transform([interacao['codigo_humano']])

    # <-- MUDAN√áA: Chamada da fun√ß√£o com os novos argumentos
    categoria_tarefa = classificar_tarefa(prompt_embedding, matriz_centroides, chaves_categorias)
    print(f"Tarefa classificada como: {categoria_tarefa}")

    iqa, churn, ss = calcular_iqa(
        interacao['codigo_ia'], interacao['codigo_humano'],
        embedding_ia, embedding_humano
    )
    print(f"Code Churn: {churn:.2f} | Similaridade Sem√¢ntica: {ss:.2f} | IQA Final: {iqa:.2f}")

    resultados.append({
        "categoria": categoria_tarefa[0],
        "complexidade": categoria_tarefa[1],
        "iqa": iqa
    })
print("\n‚úÖ Processamento conclu√≠do.\n")
```

```python
# --- Bloco de C√°lculo (seu c√≥digo original) ---
print("üìä Gerando relat√≥rio anal√≠tico do IFOR-GenAI...\n")

df_resultados = pd.DataFrame(resultados)
mapa_foi = df_resultados.groupby(['categoria', 'complexidade'])['iqa'].mean().unstack()

LIMITE_IDEAL = 0.70
total_celulas = mapa_foi.notna().sum().sum()
celulas_ideais = (mapa_foi >= LIMITE_IDEAL).sum().sum()
rangeabilidade = celulas_ideais / total_celulas if total_celulas > 0 else 0

# ==============================================================================
# NOVO BLOCO: L√ìGICA DE AN√ÅLISE AUTOM√ÅTICA
# Aqui, preparamos as frases e insights antes de montar a f-string final.
# ==============================================================================

# 1. An√°lise qualitativa da Rangeabilidade
if rangeabilidade >= 0.6:
    analise_rangeabilidade = "Excelente! A IA demonstra alta versatilidade, sendo confi√°vel na maioria das tarefas monitoradas."
elif rangeabilidade >= 0.25:
    analise_rangeabilidade = "Bom. A IA atua como uma copiloto confi√°vel em um conjunto significativo de tarefas, indicando um bom potencial de produtividade."
else:
    analise_rangeabilidade = "Ponto de Aten√ß√£o. A IA tem um papel mais focado como assistente inicial, sendo confi√°vel apenas em um n√∫mero limitado de cen√°rios."

# 2. Identifica√ß√£o dos pontos fortes e fracos
# Usamos o m√©todo stack() para facilitar a busca pelo m√°ximo e m√≠nimo no DataFrame
mapa_flat = mapa_foi.stack().dropna()

if not mapa_flat.empty:
    ponto_forte_idx = mapa_flat.idxmax()
    max_iqa = mapa_flat.max()
    texto_ponto_forte = f"O ponto de maior efic√°cia da IA foi em '{ponto_forte_idx[0]}' de complexidade '{ponto_forte_idx[1]}', com um IQA m√©dio de {max_iqa:.2f}."

    ponto_fraco_idx = mapa_flat.idxmin()
    min_iqa = mapa_flat.min()
    texto_ponto_fraco = f"A principal √°rea que requer aten√ß√£o e revis√£o cuidadosa s√£o tarefas de '{ponto_fraco_idx[0]}' de complexidade '{ponto_fraco_idx[1]}', com um IQA m√©dio de apenas {min_iqa:.2f}."

    # 3. Gera√ß√£o de recomenda√ß√µes din√¢micas
    recomendacao = f"""
    A√ß√µes Recomendadas:
    1. FOCO NO PONTO FORTE: Potencialize o uso da IA para tarefas de '{ponto_forte_idx[0]}', disseminando as melhores pr√°ticas de prompts para esta √°rea.
    2. GEST√ÉO DE RISCO: Para tarefas como '{ponto_fraco_idx[0]}', especialmente em alta complexidade, implemente uma pol√≠tica de revis√£o de c√≥digo (code review) obrigat√≥ria para as sugest√µes da IA.
    3. TREINAMENTO: Use este painel para guiar o treinamento cont√≠nuo da equipe, focando em como extrair valor mesmo nas √°reas de menor IQA.
    """
else:
    texto_ponto_forte = "N√£o h√° dados suficientes para identificar o ponto mais forte."
    texto_ponto_fraco = "N√£o h√° dados suficientes para identificar a principal √°rea de melhoria."
    recomendacao = "Colete mais dados para gerar recomenda√ß√µes acion√°veis."

# ==============================================================================
# F-STRING FINAL: O RELAT√ìRIO COMPLETO E CONTEXTUALIZADO
# ==============================================================================

relatorio_final = f"""
==================================================================
        RELAT√ìRIO DE PERFORMANCE DA IA GENERATIVA (IFOR-GenAI)
==================================================================

CONTEXTO:
Este relat√≥rio analisa a performance da ferramenta de IA com base em
intera√ß√µes reais de desenvolvimento. O objetivo √© fornecer insights
acion√°veis sobre onde a IA √© mais eficaz e onde requer supervis√£o.
Data da An√°lise: {pd.Timestamp.now().strftime('%d/%m/%Y')}

------------------------------------------------------------------
          1. MAPA DA FAIXA DE OPERA√á√ÉO IDEAL (FOI)
------------------------------------------------------------------
(O mapa abaixo mostra o √çndice de Qualidade Avan√ßado (IQA) m√©dio.
Um IQA pr√≥ximo de 1.0 indica alta qualidade e pouca edi√ß√£o humana.)

{mapa_foi.to_string(float_format="%.2f")}

------------------------------------------------------------------
        2. AN√ÅLISE DO √çNDICE DE RANGEABILIDADE (R)
------------------------------------------------------------------
Com um limiar de exig√™ncia de {LIMITE_IDEAL:.0%}, que define uma tarefa como
'confi√°vel', o √çndice de Rangeabilidade (R) da IA √© de: {rangeabilidade:.2%}

AN√ÅLISE: {analise_rangeabilidade}

Isto significa que a IA atende ao nosso padr√£o de alta qualidade em
{celulas_ideais} das {total_celulas} categorias de tarefas monitoradas.

------------------------------------------------------------------
                  3. DIAGN√ìSTICO DETALHADO
------------------------------------------------------------------
PONTO FORTE:
{texto_ponto_forte}

PONTO FRACO:
{texto_ponto_fraco}

------------------------------------------------------------------
              4. RECOMENDA√á√ïES ESTRAT√âGICAS
------------------------------------------------------------------
{recomendacao}
==================================================================
"""

# Finalmente, imprimimos o relat√≥rio completo
print(relatorio_final)
```
